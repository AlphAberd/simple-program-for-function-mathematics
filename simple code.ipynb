# <h2 id="escape">Escape Sequences</h2>

Escape sequences ditandai dengan adanya *back slash* (/) di bagian depan huruf. Escape sequences berguna untuk merepresentasikan huruf yang sulit untuk diketik. Seperti misalnya *Enter*, untuk menambahkan *Enter* (baris baru) pada sebuah kalimat dapat menggunakan back slash diikuti dengan huruf n (/n) :

```python
# Baris baru escape sequences

print(" Michael Jackson \n is the best" )
```

back slash ditambah t (/t) menghasilkan tab: 

```python
# Tab escape sequence

print(" Michael Jackson \t is the best" )
```

Jika anda ingin menuliskan back slash pada string anda, gunakanlah back slash dobel (//) :

```python
# menambahkan back slash dalam string

print(" Michael Jackson \\ is the best" )
```

kita juga bisa menambahkan r sebelum string untuk menampilkan backslash di dalam kalimat:

```python
# r akan mengatakan pada python bahwa string akan ditampilkan apa adanya (raw string).

print(r" Michael Jackson \ is the best" )
```

# <h2 id="operations">String Operations</h2>

Terdapat banyak sekali method operasi string pada python yang dapat digunakan untuk memanipulasi data. Kita akan menggunakan beberapa operasi string dasar.

Mari coba <code>upper</code>; ini adalah method yang digunakan untuk mengkonversi huruf kecil (lower case) menjadi huruf besar (upper case):

```python
# Mengonversi semua huruf menjadi huruf besar.

A = "Thriller is the sixth studio album"
print("sebelum dikapitalkan:", A)
B = A.upper()
print("setelah dikapitalkan:", B)
```

Method <code>replace</code> akan mengganti bagian dari string. Misalnya kita memiliki string baru, untuk mengganti kata pada string, masukkan kata yang ingin diganti pada argumen pertama, dan kata penggantinya pada argumen kedua :


```python
# Ganti kalimat lama jadi kalimat baru dengan mengubah satu kata pada kalimat yang lama.

A = "Michael Jackson is the best"
B = A.replace('Michael', 'Janet')
B
```

Method <code>find</code> akan mencari substring. Argumen diisi dengan kata yang ingin anda cari. Pada contoh di bawah ini, kita akan mencari tulisan <code>jack</code> atau <code>el</code>. Hasil dari method ini adalah angka yang menunjukkan lokasi index huruf pertama kali ditemukan.  

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%201/Images/StringsFind.png" width="600" align="center" />

```python
# Mencari substring dalam string. Hanya index dari elemen pertama substring yang dicetak

Name = "Michael Jackson"
Name.find('el')
```

```python
# Mencari substring dari string

Name.find('Jack')
```

Jika substring tidak ditemukan di dalam string, output program adalah (-1). Contohnya jika anda mencari 'Jasdfasdasdf' pada "Michael Jackson", substring tidak ditemukan:

```python
# Jika substring tidak ditemukan dalam string

Name.find('Jasdfasdasdf')
```
