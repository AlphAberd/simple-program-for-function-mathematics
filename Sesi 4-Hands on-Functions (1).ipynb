# <h2 id="func">Fungsi</h2>

Fungsi adalah kumpulan dari kode yang dapat digunakan kembali untuk menjalankan operasi yang ada di dalamnya. Fungsi dapat menyelesaikan permasalahan yang ada dan dapat digunakan kembali di program yang berbeda.

Terdapat dua jenis fungsi:

- <b>Fungsi yang sudah ditetapkan sebelumnya (build in)</b>
- <b>Fungsi yang ditetapkan oleh user (user defined)</b>

## <h3 id="content">Apa itu Fungsi?</h3>

Kita dapat mendefinisikan sebuah fungsi sesuai dengan kebutuhan yang diperlukan. Berikut adalah beberapa aturan sederhana yang perlu diperhatikan untuk mendefinisikan sebuah fungsi:
-  Membuat fungsi dapat dimulai dari <code>def</code>Â dan diikuti dengan <code>nama</code> fungsi serta tanda kurung <code>()</code>.
-  Tanda kurung di dalam fungsi dapat diisi dengan beberapa parameter input dan argumen. 
-  Anda juga dapat mendefinisikan beberapa parameter di dalam tanda kurung tersebut.
-  Bagian dari fungsi akan diawali dengan titik dua (<code>:</code>) dan sisi nya menjorok ke dalam.
-  Anda juga dapat menyimpan data (dokumentasi) di sebelum bagian body.
-  Pernyataan <code>return</code> digunakan secara opsional saat keluar dari fungsi, exits a function, berfungsi untuk mengembalikan nilai dari fungsi tersebut. 

Salah satu contoh fungsi yang menambahkan parameter <code>a</code> yang akan menampilkan dan mengembalikan keluarannya sebagai <code>b</code>, adalah sebagai berikut:

```python
# Contoh pertama mengenai Fungsi: menambahkan nilai 1 ke a dan disimpan sebagai b

def add(a):
    b = a + 1
    print(a, "if you add one", b)
    return(b)
```

Kita dapat membuat fungsi yang berbeda-beda. Sebagai contoh, kita dapat membuat sebuah fungsi yang dapat mengalikan dua angka. Angka-angka tersebut akan di simpan ke dalam variabel <code>a</code> dan <code>b</code>:

```python
# Mendefinisikan sebuah fungsi untuk mengalikan 2 angka

def Mult(a, b):
    c = a * b
    return(c)
```

Fungsi yang sama juga dapat digunakan untuk tipe data yang berbeda. Contoh nya, kita dapat mengalikan 2 integer sebagai berikut:

```python
# Menggunakan mult() untuk mengalikan dua tipe data yang berbeda bersamaan.

Mult(2, "Michael Jackson ")
```

## <h3 id="var">Variabel</h3>

Inputan dalam sebuah fungsi biasa disebut dengan formal parameter.

Sebuah variabel yang dideklarasi di dalam sebuah fungsi disebut dengan variabel lokal. Sebuah parameter hanya ada dalam suatu fungsi.(i.e. titik dimana fungsi dapat dimulai dan berhenti).  

Sebuah variabel yang dideklarasi di luar suatu fungsi disebut dengan global variabel, nilai dari variabel tersebut dapat di akses dan dimodifikasi sepanjang program. Kita akan mendiskusikan mengenai variabel global diakhir lab ini. 

```python
# Mendefinisikan sebuah fungsi

def square(a):
    
    # Local variable b
    b = 1
    c = a * a + b
    print(a, "if you square + 1", c) 
    return(c)
```

## <h3 id="simple">Fungsi Membuat Segalanya Mudah</h3>

Pikirkanlah 2 baris kode di dalam <b>Block 1</b> dan <b>Block 2</b>: prosedur dari setiap blok harus identik. Satu-satunya hal yang berbeda dari setiap blok adalah nama dan nilai variabel.

<h4>Blok 1:</h4>

```python
# perhitungan a dan b di blok 1

a1 = 4
b1 = 5
c1 = a1 + b1 + 2 * a1 * b1 - 1
if(c1 < 0):
    c1 = 0 
else:
    c1 = 5
c1   
```

<h4>Blok 2:</h4>

```python
# perhitungan a dan b di blok2

a2 = 0
b2 = 0
c2 = a2 + b2 + 2 * a2 * b2 - 1
if(c2 < 0):
    c2 = 0 
else:
    c2 = 5
c2   
```

Kita dapat merubah setiap line di dalam kode tersebut dengan sebuah fungsi. Sebuah fungsi, terdiri dari beberapa intruksi yang digabungkan menjadi satu. Ketika fungsi didefinisikan, fungsi tersebut dapat digunakan secara berulang kali. Anda dapat menggunakan fungsi yang sama berkali-kali d dalam suatu program. Anda juga dapat menyimpan dan menggunakan fungsi yang anda buat di program lainnya ataupun di program orang lain. Baris-baris kode yang ada di dalam kode <b>Block 1</b> dan kode <b>Block 2</b> dan diubah dengan fungsi sebagai berikut:  

```python
# Membuat sebuah fungsi untuk perhitungan berikut

def Equation(a,b):
    c = a + b + 2 * a * b - 1
    if(c < 0):
        c = 0 
    else:
        c = 5
    return(c) 
```

Fungsi ini mengambil dua input, a dan b, lalu akan dilakukan beberapa operasi dan hasilnya akan dikembaikan ke variabel c. 
Kita akan mendefinisikan sebuah fungsi sederhana, mengubah intruksi-intruksi tersebut menjadi sebuah fungsi, dan memasukan nilai baru ke dalam variabel  <code>a1</code>, <code>b1</code> dan <code>a2</code>, <code>b2</code> as sebagai sebuah inputan. Proses secara keseluruhan dapat digambarkan dalam gambar berikut ini:

```python
a1 = 4
b1 = 5
c1 = Equation(a1, b1)
c1
```

# <h2 id="if">Menggunakan pernyataan dan pengulangan <code>if</code>/<code>else</code> di dalam sebuah fungsi</h2>

Fungsi <code>return()</code> secara khusus biasa digunakan ketika anda menggunakan pernyataan IF di dalam sebuah fungsi,fungsi sangat berguna jika Anda memiliki pernyataan IF dalam fungsi, khusus nya ketika output yang ingin Anda hasilnya bergantung pada beberapa kondisi yang ada: 

```python
# Contoh sebuah fungsi

def type_of_album(artist, album, year_released):
    
    print(artist, album, year_released)
    if year_released > 1980:
        return "Modern"
    else:
        return "Oldie"
    
x = type_of_album("Michael Jackson", "Thriller", 1980)
print(x)
```

Kita dapat menggunakan pengulangan di dalam fungsi. Contohnya, kita menggunakan fungsi <code>print</code> untuk menampilkan setiap elemen di dalam list:

```python
# Menampilkan isi list dengan pengulangan

def PrintList(the_list):
    for element in the_list:
        print(element)
```

```python
# Mengimplementasikan fungsi printlist 
PrintList(['1', 1, 'the man', "abc"])
```

# <h2 id="default">Mengatur nilai tetap argumen di setiap fungsi yang dibuat</h2>

Anda bisa mengatur nilai dari argumen di dalam sebuah fungsi. Contohnya, di dalam fungsi <code>isGoodRating()</code> bagaimana jika kita ingin membuat sebuah threshold untuk apa yang dianggap mempunyai rating yan bagus. Dengan konsep default, kita akan menetapkan rating default nya 4:

```python
# Contoh konfigurasi parameter dengan nilai default

def isGoodRating(rating=4): 
    if(rating < 7):
        print("this album sucks it's rating is",rating)
        
    else:
        print("this album is good its rating is",rating)
```

# <h2 id="global">Variabel Global</h2>

Sejauh ini, kita sudah membuat beberapa variabel dengan menggunakan fungsi, tetapi kita belum mendiskusikan variabel-variabel yang berada di luar fungsi. Variabel variabel itu disebut dengan variabel global
<br>
Di sini kita akan melihat apa yang akan dikembalikan dari <code>printer1</code>:

```python
# Contoh penggunaan variabel global

artist = "Michael Jackson"
def printer1(artist):
    internal_var = artist
    print(artist, "is an artist")
    
printer1(artist)
```

Jika kita menampilkan <code>internal_var</code> kita akan mendapatkan error.

<b>Kita mendapatkan error:  <code>name 'internal_var' is not defined</code>. Kenapa?</b>  

Itu karena variabel-variabel yang dibuat di dalam sebuah fungsi merupakan <b>variabel global</b>, itu artinya, variabel tersebut tidak dapat digunakan di luar fungsi.

Tetapi, terdapat sebuah cara untuk membuat <b>variabel global</b> dari dalam sebuah fungsi seperti berikut ini:

```python
artist = "Michael Jackson"

def printer(artist):
    global internal_var 
    internal_var= "Whitney Houston"
    print(artist,"is an artist")

printer(artist) 
printer(internal_var)
```

<h2 id="scope">Ruang Lingkup Variabel</h2>

Lingkup variabel adalah bagian dari sebuah program dimana variabel tersebut dapat diakses secara umum. Variabel-variabel tersebut di deklarasikan di lluar fungsi yang didefinisikan, seperti variabel <code>myFavouriteBand</code> yang ditampilkan di kode berikut, merupakan variabel yang dapat diakses secara umum di dalam program ini. Singkatnya, beberapa variabel tersebut yang mempunyai lingkup global, termasuk ke dalam variabel global.

<code>myFavouriteBand</code> adalah sebuah variabel global, maka dapat diakses di dalam fungsi <code>getBandRating</code>, dan kita dapat menggunakan variabel tersebut untuk menentukan rating dari sebuah band. Kita juga dapat menggunakan variabel tersebut di luar sebuah fungsi, yang mana kita akan meneruskannya ke dalam fungsi print untuk menampilkannya:

```python
# Contoh penggunaan global variabel

myFavouriteBand = "AC/DC"

def getBandRating(bandname):
    if bandname == myFavouriteBand:
        return 10.0
    else:
        return 0.0

print("AC/DC's rating is:", getBandRating("AC/DC"))
print("Deep Purple's rating is:",getBandRating("Deep Purple"))
print("My favourite band is:", myFavouriteBand)
```

Berikut merupakan modifikasi dari kode yang sudah dibuat di atas. Sekarang, variabel <code>myFavouriteBand</code> sudah didefinisikan di dalam fungsi <code>getBandRating</code>. Variabel yang di definisikan di dalam suatu fungsi disebut dengan lokal varibel. Artinya, variabel tersebut hanya diakses di dalam fungsi tersebut. Fungsi <code>getBandRating</code> akan tetap berjalan, dikarenakan variabel <code>myFavouriteBand</code> masih terdefinisi di dalam fungsi tersebut. Namun, kita tidak dapat menampilkan kembali nilai variabel di luar fungsi tersebut, hal ini dikarenakan variabel tersebut merupakan lokal varibael dari fungsi <code>getBandRating</code>; yang mana variabel tersebut hanya didefinisikan di dalam fungsi <code>getBandRating</code>:

```python
# Contoh penggunaan global variabel

def getBandRating(bandname):
    myFavouriteBand = "AC/DC"
    if bandname == myFavouriteBand:
        return 10.0
    else:
        return 0.0

print("AC/DC's rating is: ", getBandRating("AC/DC"))
print("Deep Purple's rating is: ", getBandRating("Deep Purple"))
print("My favourite band is", myFavouriteBand)
```

Terakhir, kita akan melihat contoh berikut. Kita akan mendefinisikan dua variabel <code>myFavouriteBand</code>. Salah satu variabel memiliki lingkup global, dan yang satu nya memiliki lingkup lokal di dalam fungsi <code>getBandRating</code>. Di dalam fungsi <code>getBandRating</code>, lokal variabel akan lebih dahulu dijalankan. **Deep Purple** akan menerima rating 10.0 setelah melewati fungsi <code>getBandRating</code>. Namun, di luar dari fungsi <code>getBandRating</code>, variabel lokal dari fungsi  <code>getBandRating</code> tidak akan terdefinisikan, oleh karena itu variabel <code>myFavouriteBand</code> yang akan ditamppilkan merupakan variabel global, yang mempunyai nilai **AC/DC**:

```python
# Contoh penggunaan global variabel dan lokal variabel denan nama yang sama

myFavouriteBand = "AC/DC"

def getBandRating(bandname):
    myFavouriteBand = "Deep Purple"
    if bandname == myFavouriteBand:
        return 10.0
    else:
        return 0.0

print("AC/DC's rating is:",getBandRating("AC/DC"))
print("Deep Purple's rating is: ",getBandRating("Deep Purple"))
print("My favourite band is:",myFavouriteBand)
```
